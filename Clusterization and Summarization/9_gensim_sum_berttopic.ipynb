#### Implement

```python
tokenizer = Tokenizer('english')

def summarize_func(text, summarizer, sentences_count=1):
  parser = PlaintextParser.from_string(text,tokenizer)
  summary = summarizer(parser.document,sentences_count=sentences_count)
  return summary

def run_summarization(df, summarizer, verbose=False, sentences_count=1):
  df_new = df.copy()
  df_new['kp'] = 'no'
  res = {}
  for i,group in df.groupby(['name','cluster_id']):
    args = [process_arg(arg) for arg in group['argument'].tolist()]
    text = '. '.join(args)
    ids = group['id'].tolist()
    sum = summarize_func(text,summarizer,sentences_count=sentences_count)
    res[i]=sum
    if verbose:
      print(i)
      print('IDS')
      print(ids)
      print('ARGS LEN')
      print(len(args))
      print('ARGS')
      print(args)
      print('TEXT')
      print(text)
      print('SUM')
      print(sum)
    for id in ids:
      df_new.loc[df_new['id']==id, 'kp']=str(sum[0]) 
  return df_new
```

#### Run lex rank

```python
lex_rank_summarizer = LexRankSummarizer()
df_lex_rank = run_summarization(df,lex_rank_summarizer)
df_lex_rank.head(2)
```

```python
def clear_arg(arg):
  arg = re.sub(r"\r","",arg)
  arg = re.sub(r"\n"," ",arg)

  arg = re.sub(r"\(Art\. .*\)","ARTICLE",arg)
  arg = re.sub(r"\(art\. .*\)","ARTICLE",arg)
  # arg = re.sub(r"\(Art\. [0-9]+\)","ARTICLE",arg)
  arg = re.sub(r"No\.","number ",arg)
  arg = re.sub(r"no\.","number ",arg)
  arg = re.sub(r"Eur\.","European",arg)
  arg = re.sub(r"eur\.","European",arg)
  arg = re.sub(r"P\.","page ",arg)
  arg = re.sub(r"p\.","page ",arg)
  arg = re.sub(r"Nos\.","not otherwise specified",arg)
  arg = re.sub(r"nos\.","not otherwise specified",arg)
  arg = re.sub(r"para\.","paragraph",arg)
  arg = re.sub(r"cf\.","LINK",arg)
  arg = re.sub(r"[0-9]+","INT",arg)
  #end
    # omit par. or
  arg = arg.lstrip(" (")
  arg = arg.rstrip(" ).")

  arg = re.sub(r"  "," ",arg)
  arg = re.sub(r" \.",".",arg)
  return arg
  # arg = arg.replace
print(clear_arg('\r\r\rlololol\r\r\r pp ppp\n\n\n'))
print(clear_arg(df['argument'].tolist()[i]))
```

```python
from transformers import pipeline

# summarizer = pipeline("summarization")
# summarizer = pipeline("summarization", model="facebook/bart-large-cnn", tokenizer="facebook/bart-large-cnn")
summarizer = pipeline("summarization", model="nsi319/legal-pegasus", tokenizer="nsi319/legal-pegasus")
```

```python
def run_summarization_transformer(df, verbose=False):
  df_new = df.copy()
  df_new['kp'] = 'no'
  res = {}
  for i,group in df.groupby(['name','cluster_id']):
    args = [process_arg(arg) for arg in group['argument'].tolist()]
    text = '. '.join(args)
    ids = group['id'].tolist()
    if verbose:
      print(i)
      print('IDS')
      print(ids)
      print('ARGS LEN')
      print(len(args))
      print('ARGS')
      print(args)
      print('TEXT')
      print(text)
    sum = summarizer(text, min_length=1, max_length=10)
    res[i]=sum
    if verbose:
      print('SUM')
      print(sum)
    for id in ids:
      df_new.loc[df_new['id']==id, 'kp']=str(sum[0]) 
  return df_new
```
